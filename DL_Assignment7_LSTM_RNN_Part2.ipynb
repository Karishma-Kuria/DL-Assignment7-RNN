## **1. Many-to-One Sequence Problems**

### In many-to-one sequence problems, we have a sequence of data as input and we have to predict a single output. Text classification is a prime example of many-to-one sequence

### **Create Database**

Creating a list of integers from 1 to 45. In further steps I want 15 samples in the dataset, I will reshape the list of integers containing the first 45 integers.

```python
X_arr = np.array([x+1 for x in range(45)])
print(X_arr)
```

Now lets reshape the input into time steps, features, number of samples.

```python
X = X_arr.reshape(15,3,1)
print(X_arr)
```

Now the input is converted into 3D vector. Since every element in the output will be equal to the sum of the values in the time-steps in the corresponding input sample. The following script will create the output vector.

```python
Y = list()
for x in X:
    Y.append(x.sum())

Y = np.array(Y)
print(Y)
```

### **Solution via Simple LSTM**

```python
# create model with one layer
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(3, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

#### **Train the model**

```python
history = model.fit(X, Y, epochs=1000, validation_split=0.2, verbose=1)
```

Lets test the trained model for sequence 52,53,54. The actual output should be 52+53+54 = 159

```python
test_input = array([52,53,54])
test_input = test_input.reshape((1, 3, 1))
test_output = model.predict(test_input, verbose=0)
print(test_output)
```

### The output is which is  points less than 159

### **Solution via Stacked LSTM**

Lets try with multiple LSTM layers.

```python
model = Sequential()
model.add(LSTM(200, activation='relu', return_sequences=True, input_shape=(3, 1)))
model.add(LSTM(100, activation='relu', return_sequences=True))
model.add(LSTM(50, activation='relu', return_sequences=True))
model.add(LSTM(25, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

history = model.fit(X, Y, epochs=1000, validation_split=0.2, verbose=1)
```

#### Lets test the trained model for sequence 52,53,54.

```python
test_output = model.predict(test_input, verbose=0)
print(test_output)
```

#### The output is which is 155.38 which is approx 4 less than 159.

### **Solution via Bidirectional LSTM**

Lets try Bidirectional LSTM to further test the results.

```python
from keras.layers import Bidirectional

model = Sequential()
model.add(Bidirectional(LSTM(50, activation='relu'), input_shape=(3, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

#### Train and test for input 52,53,54

```python
history = model.fit(X, Y, epochs=1000, validation_split=0.2, verbose=1)
test_output = model.predict(test_input, verbose=0)
print(test_output)
```

#### The output is which is 158.70 very close to 159, this ouput is better than the single layer and stacked unidirectional LSTMs approach we used for this dataset.

## **2. One-to-many:**  

### sequence problems are sequence problems where the input data has one time-step, and the output contains a vector of multiple values or multiple time-steps. Thus, we have a single input and a sequence of outputs.


### **Create the Dataset**

```python
from numpy import array
from keras.preprocessing.text import one_hot
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers.core import Activation, Dropout, Dense
from keras.layers import Flatten, LSTM
from keras.layers import GlobalMaxPooling1D
from keras.models import Model
from keras.layers.embeddings import Embedding
from sklearn.model_selection import train_test_split
from keras.preprocessing.text import Tokenizer
from keras.layers import Input
from keras.layers.merge import Concatenate
from keras.layers import Bidirectional

import pandas as pd
import numpy as np
import re

import matplotlib.pyplot as plt

X = list()
Y = list()
X = [x+3 for x in range(-2, 43, 3)]

for i in X:
    output_vector = list()
    output_vector.append(i+1)
    output_vector.append(i+2)
    Y.append(output_vector)

print(X)
print(Y)
```

Our input contains 15 samples with one time-step and one feature value. For each value in the input sample, the corresponding output vector contains the next two integers.

Lets reshape the data as per thhe LSTM model.

```python
X = np.array(X).reshape(15, 1, 1)
Y = np.array(Y)
```

### **Solution via Simple LSTM**

```python
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(1, 1)))
model.add(Dense(2))
model.compile(optimizer='adam', loss='mse')
model.fit(X, Y, epochs=1000, validation_split=0.2, batch_size=3)
```

Once the model is trained, lets do prediction for one of the values.

#### I have 13 in input so the output should be 14 and 15.

```python
test_input = array([13])
test_input = test_input.reshape((1, 1, 1))
test_output = model.predict(test_input, verbose=0)
print(test_output)
```

So as a result of prediction I got values [14.030346 15.020935] which is very clode to 14 and 15.

### **Solution via Stacked LSTM**

#### Lets increase number of layers using stacked LSTM

```python
model = Sequential()
model.add(LSTM(50, activation='relu', return_sequences=True, input_shape=(1, 1)))
model.add(LSTM(50, activation='relu'))
model.add(Dense(2))
model.compile(optimizer='adam', loss='mse')
```

#### **Train the model**

```python
history = model.fit(X, Y, epochs=1000, validation_split=0.2, verbose=1, batch_size=3)
```

#### Lets test the output for 13.

```python
test_output = model.predict(test_input, verbose=0)
print(test_output)
```

#### So the conclusion is Single layer LSTM worked better for this data since the output with Stacked LSTM is [13.946449 14.948654] which is less closer to the expected output which is 14 and 15.

## **3. Many-to-Many Sequence Problems**


#### Many-to-Many sequence learning can be used for machine translation where the input sequence is in some language, and the output sequence is in some other language. It can be used for Video Classification as well, where the input sequence is the feature representation of each frame of the video at different time steps. 

### **Create Dataset**

```python
from numpy import array
from keras.preprocessing.text import one_hot
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers.core import Activation, Dropout, Dense
from keras.layers import Flatten, LSTM
from keras.layers import GlobalMaxPooling1D
from keras.models import Model
from keras.layers.embeddings import Embedding
from sklearn.model_selection import train_test_split
from keras.preprocessing.text import Tokenizer
from keras.layers import Input
from keras.layers.merge import Concatenate
from keras.layers import Bidirectional

import pandas as pd
import numpy as np
import re

import matplotlib.pyplot as plt

X = list()
Y = list()
X = [x for x in range(5, 301, 5)]
Y = [y for y in range(20, 316, 5)]

X = np.array(X).reshape(20, 3, 1)
Y = np.array(Y).reshape(20, 3, 1)
```

### **Solution via Stacked LSTM**

In the below script, the first LSTM layer is the encoder layer.

```python
from keras.layers import RepeatVector
from keras.layers import TimeDistributed

model = Sequential()

# encoder layer
model.add(LSTM(100, activation='relu', input_shape=(3, 1)))

# repeat vector
model.add(RepeatVector(3))

# decoder layer
model.add(LSTM(100, activation='relu', return_sequences=True))

model.add(TimeDistributed(Dense(1)))
model.compile(optimizer='adam', loss='mse')

print(model.summary())
```

#### **Train the model**

```python
history = model.fit(X, Y, epochs=1000, validation_split=0.2, verbose=1, batch_size=3)
```

#### **Create Test points**

Lets predict using test data if the encode-decoder model is able to predict.

```python
test_input = array([300, 305, 310])
test_input = test_input.reshape((1, 3, 1))
test_output = model.predict(test_input, verbose=0)
print(test_output)
```

So the output I get is in 3D form.
